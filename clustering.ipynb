```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
features = pd.read_csv('Dataset/pros_features.csv')
targets = pd.read_csv('Dataset/pros_targets.csv')
```

```python
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters = k, random_state = 42)
    kmeans.fit(features)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wcss, marker = 'o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters (k)')
plt.ylabel('WCSS')
plt.show()
```

Choosing k = 6

```python
kmeans = KMeans(n_clusters = 6, random_state = 42)
features['Clusters'] = kmeans.fit_predict(features)
```

```python
plt.scatter(features[:, 0], features[:, 1], c = features['Clusters'], cmap = 'viridis')
plt.title('Clusters')
plt.xlabel('first col')
plt.ylabel('second col')
plt.show()
```
